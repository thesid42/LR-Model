```python

import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

N = 100
v = input()
#add .csv at the end of input
with open(v, 'w') as f:
    X = np.random.uniform(low=0, high=1000, size=N)
    Y = np.random.uniform(low = 20,high = 50, size=N)
    for i in range(N):
        f.write("%s,%s\n" % (X[i], Y[i]))
```

```python
rpm = []
torque = []
for line in open(v):
    x, y = line.split(',')
    x= float(x)
    y = float(y)
    x= round(x,2)
    y =round(y,2)
    rpm.append(x)
    torque.append(y)

rpm = np.array(rpm)
torque = np.array(torque)
```

```python
for a in range(0,len(X)):
    torque[i] = float(torque[i])
    torque[i] = round(torque[i],2)
    torque[i] = torque[i]*9.807
```

```python
energy=[]
i=0
j=0
while i<len(rpm) and j<len(torque):
    a  = (rpm[i]*torque[j])/9.5488
    a = float(a)
    a = round(a,2)
    energy.append(a)
    i+=1
    j+=1
    
energy = np.array(energy)

with open('new.csv', 'w') as f:
    for i in range(N):
        f.write("%s,%s,%s\n" % (rpm[i], torque[i],energy[i]))
    
```
